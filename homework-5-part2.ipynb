```python
import requests
import pprint
```

```python
# Current weather I was born

url = "https://api.tomorrow.io/v4/weather/realtime?location=karachi&apikey={OPENMETEO_API_KEY}"

headers = {
    "accept-encoding": "deflate, gzip, br",
    "accept": "application/json"
}

response = requests.get(url, headers=headers)
karachi = response.json()

pprint.pp(karachi)
```

```python
# Country
print(karachi['location']['name'])
# Idk why they decided to have it be in Urdu but but the leftmost word says Pakistan trust me
```

```python
# Difference between current temp and how warm it feels
print(f'It feels {karachi['data']['values']['temperatureApparent'] - karachi['data']['values']['temperature']:1f} degrees warmer than it is.')
```

```python
# Current temp at Heathrow
url = "https://api.tomorrow.io/v4/weather/realtime?location=51.470020,-0.454295&apikey={OPENMETEO_API_KEY}"

headers = {
    "accept-encoding": "deflate, gzip, br",
    "accept": "application/json"
}

response = requests.get(url, headers=headers)
heathrow = response.json()

pprint.pp(f'The current temperature at Heathrow is {heathrow['data']['values']['temperature']} degrees.')
```
